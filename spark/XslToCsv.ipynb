```python
bucket_input = "lz-prices"
bucket_output = "lz-prices-csv"

client = storage.Client()
xlsx_files = list(client.list_blobs("lz-prices"))

for blob in xlsx_files:
    xlsx_file_path = f"gs://{bucket_input}/{blob.name}"

    # Load XLSX file as a Pandas DataFrame
    df = pd.read_excel(xlsx_file_path, engine="openpyxl", header=0)
    
    df

    # Construct the output CSV file path
    csv_filename = os.path.splitext(blob.name)[0] + ".csv"
    csv_file_path = f"gs://{bucket_output}/{csv_filename}"

    # Save the DataFrame as a CSV file in Google Cloud Storage
    df.to_csv(csv_file_path, index=False, header=True, sep = ';')

    print(f"Converted {xlsx_file_path} to {csv_file_path}")
```
